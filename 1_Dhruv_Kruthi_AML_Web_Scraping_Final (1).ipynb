# Pass 1 of Web Scraping

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# Sample: assume your dataframe is named `urls` with column 'url'
# urls = pd.DataFrame({'url': [... your list of URLs ...]})
urls = pd.read_csv("url_only_data.csv")
cnt = 0

def extract_info(row):
    global cnt  # Access the global cnt variable
    url = row['url']
    cnt += 1
    if cnt % 100 == 0:
        print(cnt)
    # Determine news source
    if "foxnews.com" in url:
        source = "Fox News"
    elif "nbcnews.com" in url:
        source = "NBC News"
    else:
        source = "Unknown"

    # Try to fetch and parse the headline
    try:
        response = requests.get(url)
        if response.status_code != 200:
            return pd.Series([source, None])

        soup = BeautifulSoup(response.text, "html.parser")

        # Fox News headlines
        if source == "Fox News":
            title_tag = soup.find("title")

        # NBC News headlines
        elif source == "NBC News":
            title_tag = soup.find("title")  # Often works; may adjust based on NBC's structure

        headline = title_tag.get_text(strip=True) if title_tag else None

    except Exception as e:
        headline = None

    return pd.Series([source, headline])

# Apply the function to each row
urls[['news', 'headline']] = urls.apply(extract_info, axis=1)

urls.head()
```

```python
test = urls[urls['headline'].isnull()]  # Works for both None and np.nan
```

```python
urls.to_csv("headlines.csv", index=False)
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time

cnt = 0

# Retry function to fetch the HTML content
def fetch_with_retries(url, headers=None, retries=3, delay=2):
    for attempt in range(retries):
        try:
            response = requests.get(url, headers=headers, timeout=10)
            if response.status_code == 200:
                return response.text
        except Exception as e:
            pass
        time.sleep(delay)  # wait before retrying
    return None  # return None after all retries fail

def extract_info(row):
    global cnt
    url = row['url']
    cnt += 1
    if cnt % 100 == 0:
        print(f"Processing row: {cnt}")

    # Determine news source
    if "foxnews.com" in url:
        source = "Fox News"
    elif "nbcnews.com" in url:
        source = "NBC News"
    else:
        source = "Unknown"

    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                      "AppleWebKit/537.36 (KHTML, like Gecko) "
                      "Chrome/122.0.0.0 Safari/537.36"
    }

    html = fetch_with_retries(url, headers=headers)
    if html is None:
        return pd.Series([source, None])  # failed to fetch

    try:
        soup = BeautifulSoup(html, "html.parser")

        # Fox News headlines
        if source == "Fox News":
            title_tag = soup.find("title")

        # NBC News headlines
        elif source == "NBC News":
            title_tag = soup.find("title")

        else:
            title_tag = soup.find("title")

        headline = title_tag.get_text(strip=True) if title_tag else None

    except Exception as e:
        headline = None

    return pd.Series([source, headline])

test2[['news', 'headline']] = test2.apply(extract_info, axis=1)

test2.head()
```

# Since headlines may have gotten the news outlet as well, we just remove everything after the '|' symbol

```python
# Create a copy of the original dataframe
cleaned_df = combined_df.copy()

# Remove everything after and including the '|' in the headline column
cleaned_df['headline'] = cleaned_df['headline'].str.replace(r'\|.*', '', regex=True).str.strip()
cleaned_df
```
